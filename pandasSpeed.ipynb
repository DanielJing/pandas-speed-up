```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('best artists_RS.csv')
df
```

```python
df.describe()
```

```python
def reputation(row):
    if row['Total Awards'] > 1:
        return print('Good')
    if row['Total Awards'] > 5:
        return print('Great')
```

## Loop

```python
%%timeit
for index, row in df.iterrows():
    df.loc[index] = reputation(row)
```

## Apply

```python
%%timeit
df2 = df.apply(reputation, axis = 1)
```
